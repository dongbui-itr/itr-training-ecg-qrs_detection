```
import pandas as pd
from typing import Optional
import plotly
import plotly.graph_objects as go


def plot_results_1d(
    x_: Optional[np.ndarray],
    y_: Optional[np.ndarray],
    y_true_: Optional[np.ndarray] = None,
    estimates_: Optional[pd.DataFrame] = None,
    title: str = "",
    estimates_mode: str = "lines+markers",
) -> None:
    """
    Plots a scatter plot of all columns in the provided dataframe where each column is treated as a separate variable.
    It is assumed estimates_.index is the same as x_ as both are used for x coordinates in the plot.

    Args:
        x_: The predictors X for all observations with shape [N,K] where N is the number of observations and K is the
            dimension of X.
        y_: The corresponding response for all observations.
        y_true_: The ground truth response at the locations x.
        estimates_: The estimates y_hat given x. it is assumed each column in the dataframe is the results of a
            different estimator.
        title: The plot title.
        estimates_mode: The style mode for estimates data.
    """
    fig = go.Figure()
    if x_ is not None and y_ is not None:
        fig.add_trace(go.Scatter(x=x_, y=y_, mode="markers", name="data"))

    if y_true_ is not None:
        fig.add_trace(
            go.Scatter(x=x_ if x_ is not None else estimates_.index, y=y_true_, mode="markers", name="y_true")
        )

    if estimates_ is not None:
        for c in estimates_.columns:
            fig.add_trace(go.Scatter(x=estimates_.index, y=estimates_[c], mode=estimates_mode, name=c))

    fig.update_layout(
        go.Layout(
            title=title,
            autosize=True,
            height=2000,
        )
    )

    plotly.offline.iplot(fig)
```

```
#180sample=0.5s
#90samples=0.25s
def _normalize_arr(y_arr):
    window_size=90
    res_arr=np.empty_like(y_arr)
    for i in range(len(y_arr)):
        lower_ind=max(i-window_size,0)
        higher_ind=min(i+window_size,len(y_arr)-1)
        # res_arr=np.abs(np.max(y_arr[lower_ind:higher_ind])),
        res_arr[i]=max(np.abs(np.max(y_arr[lower_ind:higher_ind])),np.abs(np.min(y_arr[lower_ind:higher_ind])))
    return res_arr
```

```
nor_y=_normalize_arr(y_final)
mean_nor=np.mean(nor_y)/2
# mean_nor
nor_y=np.where(nor_y<mean_nor,mean_nor,nor_y)
y_final_norm=(y_final/nor_y)
plot_results_1d(
    x_=x_ecg[:3000],
    y_=y_ecg[:3000],
    y_true_=y_final_norm[:3000]*50+800,#scale for easier tracking data 
    estimates_=pd.DataFrame(data=y_final_norm[:3000]*50+800, columns=["y_final_norm"], index=x_ecg[:3000]),
    title="Normalization signal",
)
```

```
def labelingForY(x_train,y_train):
    #relabelling for each data
    #positive if inrange of 40ms(15 point bef & after)
    #to category the class (1 hot encoder)
    y_train_final=np.zeros(x_train.shape[0])
    y_train_final=y_train_final.transpose()
    myData=zip(y_train['labels'],y_train['sample_dot'])
    for i in myData:
        if i[0]=='N':
            for j in range(-15,16):
                if 0<i[1]+j<650000:
                   y_train_final[i[1]+j]=1
    y_train_final=to_categorical(y_train_final)
    return y_train_final

def resamplingData(x_train):
    #each data point represent for 36 point before and 109 point after
    new_arr=np.pad(x_train,(36,109),'mean')
    resamplingXTrain=np.empty([x_train.shape[0],145])
    for i in range(x_train.shape[0]):
        for j in range(i,i+145):
            resamplingXTrain[i,j-i]=new_arr[j]
    return resamplingXTrain
```

```
from tensorflow.keras.utils import to_categorical
x_dat=resamplingData(y_final_norm)
y_dat=labelingForY(y_final_norm,test_df)
```

```
x_dat=x_dat.reshape(x_dat.shape[0],145,1)
x_train=x_dat[:350000]
x_test=x_dat[350000:]
y_train=y_dat[:350000]
y_test=y_dat[350000:]
```

AI model

```
import tensorflow as tf
from tensorflow.keras import layers, models


input_shape = (145,1)

# Define the model
model = models.Sequential()

# Input layer (Implicit when defining the input shape for the first layer)
model.add(layers.InputLayer(input_shape=input_shape))

# First convolutional layer
model.add(layers.Conv1D(filters=32, kernel_size=5, strides=1, activation='relu'))
model.add(layers.Dropout(0.5))

# First pooling layer
model.add(layers.MaxPooling1D(pool_size=3, strides=2))

# Second convolutional layer
model.add(layers.Conv1D(filters=32, kernel_size=5, strides=1, activation='relu'))
model.add(layers.Dropout(0.5))

# Flatten the output for the fully connected layers
model.add(layers.Flatten())

# First fully connected layer
model.add(layers.Dense(1024, activation='relu'))
model.add(layers.Dropout(0.5))

# Second fully connected layer
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dropout(0.5))

# Output layer
model.add(layers.Dense(2, activation='softmax'))

# Print the model summary
model.summary()

# Compile the model
#learn 
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.05, momentum=0.9),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```
# Print the model summary
model.summary()
# y_test=y_train[:14]
# x_train
# Train the model
history = model.fit(x_train, y_train, epochs=3, batch_size=128, validation_data=(x_test,y_test))

# Evaluate the model on the validation set
val_loss, val_acc = model.evaluate(x_test, y_test, verbose=2)
print(f"Validation Accuracy: {val_acc}")
```
